```python
for col in df.columns:
    if df[col].dtype== 'object':
        df[col]=df[col].str.replace(',','.')
    try:
        df[col]=df[col].astype(float)
    except:
        ValueError
    pass
```

```python
df.dtypes
```

```python
df['Service'].unique()
```

```python
df['GDP ($ per capita)']=df['GDP ($ per capita)'].fillna(df['GDP ($ per capita)'].mean())
df['Net migration']=df['Net migration'].fillna(df['Net migration'].mean())
df['Literacy (%)']=df['Literacy (%)'].fillna(df['Literacy (%)'].mean())
df['Phones (per 1000)']=df['Phones (per 1000)'].fillna(df['Phones (per 1000)'].mean())
df['Arable (%)']=df['Arable (%)'].fillna(df['Arable (%)'].mean())
df['Crops (%)']=df['Crops (%)'].fillna(df['Crops (%)'].mean())
df['Other (%)']=df['Other (%)'].fillna(df['Other (%)'].mean())
df['Climate']=df['Climate'].fillna(df['Climate'].mode()[0])
df['Birthrate']=df['Birthrate'].fillna(df['Birthrate'].mean())
df['Deathrate']=df['Deathrate'].fillna(df['Deathrate'].mean())
df['Agriculture']=df['Agriculture'].fillna(df['Agriculture'].mean())
df['Industry']=df['Industry'].fillna(df['Industry'].mean())
df['Service']=df['Service'].fillna(df['Service'].mean())
df['Infant mortality (per 1000 births)']=df['Infant mortality (per 1000 births)'].fillna(df['Infant mortality (per 1000 births)'].mean())
```

```python
df.isna().sum()
```

```python
sns.boxplot(y='Service',data=df)
```

```python
columns = ['GDP ($ per capita)', 'Literacy (%)', 'Birthrate', 'Deathrate', 'Agriculture']
for col in columns:
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1

    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR

    df = df[(df[col]>=lower_bound) & (df[col]<=upper_bound)]
    sns.boxplot(df[col])
    plt.title('Boxplot w/o outliers')
    plt.show()

```

```python
x=df.drop(['Country','Region'],axis=1)
x
```

```python

from sklearn.cluster import KMeans
inertia=[]
for i in range(1,11):
    data=KMeans(n_clusters=i,init='k-means++',random_state=42)
    data.fit(x)
    inertia.append(data.inertia_)
```

```python
print(inertia)
```

```python
plt.plot(range(1,11),inertia)
plt.xlabel('No. of clusters')
plt.ylabel('Variance')
```

```python
data1=KMeans(n_clusters=3,init='k-means++',random_state=42)
data1.fit(x)
y_pred=data1.predict(x)
```

```python
x['Clusters']=y_pred
x
```

```python
x['Clusters'].value_counts()
```

```python
sns.histplot(x['Clusters'],kde=True)
```

```python
x1=x.iloc[:,:-1].values
y1=x.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
from sklearn.naive_bayes import MultinomialNB
nav=MultinomialNB()
from sklearn.svm import SVC
sup=SVC()
from sklearn.metrics import confusion_matrix, classification_report
lst=[knn,nav,sup]
```

```python
for i in lst:
    i.fit(x_train,y_train)
    y_pred1=i.predict(x_test)
    print('Model is:',i)
    print(confusion_matrix(y_test,y_pred1))
    print(classification_report(y_test,y_pred1))
```
